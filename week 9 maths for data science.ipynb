## 1. Finding the Eigenvalues and Eigenvectors of a Matrix

```python
#(a) Is the vector  an eigenvector of A? Verify your answer with a calculation in Python.


import numpy as np
A = np.array([[4,0,1], [-1,-6,2], [5,0,0]])

v = np.array([[1],[2],[3]])

print (A.dot(v))
print('A*v doesnot have a common factor to be a scaled down to the original v vector\n hence it is not a eigenventor.')
```

```python
#(b) Is the vector [0],[1],[0] an eigenvector of A? Verify your answer with a calculation in Python.

v1 = np.array([[0],[1],[0]])
print (A.dot(v1))
print ('A*v1 has common facor of-6 which is the eigen value\n and it cound be scaled down to original v1.It is a eigenvector.')
```

```python
# (c) Use Python to find all of the eigenvalues and eigenvectors of A.
from numpy.linalg import eig
w,v=eig(A)
print('E-value:', w)
print('E-vector', v)
```
